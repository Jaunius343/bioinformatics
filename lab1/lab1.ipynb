```python
def find_all_start_stop_pairs(codon_sequence):
    start_codon = "ATG"
    stop_codons = ["TAA", "TAG", "TGA"]

    pairs = []
    start_found = False
    start_index = -1

    for i in range(0, len(codon_sequence) - 2, 3):
        codon = codon_sequence[i:i + 3]

        if codon == start_codon:
            if not start_found:
                start_found = True
                start_index = i

        if start_found and codon in stop_codons:
            pairs.append((start_index, i))  # Add the start-stop pair
            start_found = False

    return pairs
```

```python
def construct_fragments(sequence, start_stop_pairs):
    fragments = []
    
    for start, stop in start_stop_pairs:
        fragment = sequence[start:stop + 3]  # Pridedame 3 nukleotidus po stop codono
        fragments.append(fragment)
    
    return fragments
```

```python
def find_fragments_from_sequence(sequence):
    fragments = []

    for i in range(3):
        frame = sequence[i:]
        if(debug):
            print(f"frame_{i}: {frame}")
            
        start_stop_pairs = find_all_start_stop_pairs(frame)
        fragment_from_frame = construct_fragments(frame, start_stop_pairs)
        fragments = fragments + fragment_from_frame

    return fragments
```

```python
from collections import Counter

def calculate_codon_freq(protein_sequences):
    amino_acid_counts = Counter()
    for protein_sequence in protein_sequences:
        amino_acid_counts += Counter(protein_sequence)
    
    return amino_acid_counts
```

```python
from collections import Counter

def calculate_codon_freq_normalized(protein_sequences):
    amino_acid_counts = Counter()
    total_amino_acids = 0

    # Count amino acids and calculate the total number of amino acids
    for protein_sequence in protein_sequences:
        amino_acid_counts += Counter(protein_sequence)
        total_amino_acids += len(protein_sequence)

    # Normalize the counts by dividing by the total number of amino acids
    amino_acid_freq = {aa: count / total_amino_acids for aa, count in amino_acid_counts.items()}

    return amino_acid_freq

# # Example usage:
# protein_sequences = ["MAFTY", "YKLPT", "KKKM"]
# normalized_freq = calculate_amino_acid_freq(protein_sequences)

# # Print normalized frequencies
# for amino_acid, frequency in normalized_freq.items():
#     print(f"Amino Acid: {amino_acid}, Frequency: {frequency:.4f}")
```

```python
import numpy as np

def calculate_dicodon_freq(protein_sequences):
    # Create a 20x20 matrix, where one axis corresponds to one amino acid, and the other to another amino acid
    amino_acids = "ACDEFGHIKLMNPQRSTVWY"  # 20 amino acids
    matrix = np.zeros((20, 20), dtype=int)

    # Calculate the frequency of each dicodon (two symbols at a time) in all sequences
    for protein_sequence in protein_sequences:
        for i in range(0, len(protein_sequence) - 1, 2):
            dicodon = protein_sequence[i:i+2]
            amino_acid1, amino_acid2 = dicodon[0], dicodon[1]
            row_index = amino_acids.index(amino_acid1)
            col_index = amino_acids.index(amino_acid2)
            matrix[row_index][col_index] += 1
    return matrix
```

```python
import numpy as np

def calculate_dicodon_freq_normalized(protein_sequences):
    amino_acids = "ACDEFGHIKLMNPQRSTVWY"  # 20 amino acids
    matrix = np.zeros((20, 20), dtype=int)

    # Calculate the frequency of each dicodon (two symbols at a time) in all sequences
    total_dicodons = 0  # Initialize the total dicodon count

    for protein_sequence in protein_sequences:
        for i in range(0, len(protein_sequence) - 1, 2):
            dicodon = protein_sequence[i:i+2]
            amino_acid1, amino_acid2 = dicodon[0], dicodon[1]
            row_index = amino_acids.index(amino_acid1)
            col_index = amino_acids.index(amino_acid2)
            matrix[row_index][col_index] += 1
            total_dicodons += 1

    # Normalize the matrix by dividing by the total dicodon count
    normalized_matrix = matrix / total_dicodons

    return normalized_matrix
```

### get frequencies of all sequences

```python
def calculate_frequencies(sequence_file):
    # Nuskaitome seką iš failo su fasta formatu
    sequence = SeqIO.read(sequence_file, "fasta").seq
    sequence_name = SeqIO.read(sequence_file, "fasta").id

    # surandame visus fragmentus
    fragments_start_stop= []
    fragments_start_stop = find_fragments_from_sequence(sequence)
    rev_seq = sequence.reverse_complement()
    fragments_start_stop = fragments_start_stop + find_fragments_from_sequence(rev_seq)

    # filtruojame elementus
    filtered_fragments_start_stop = filter_long_elements(fragments_start_stop)
    print(f"filtered length start-stop: {len(filtered_fragments_start_stop)}") 

    # transliuojam visus codonus į baltymų seką
    translated_fragments_start_stop = [element.translate() for element in filtered_fragments_start_stop]

    # pašalinam terminatin * simbolį
    translated_fragments_start_stop = [element[:-1] for element in translated_fragments_start_stop]
    # print(translated_fragments_start_stop)

    codon_freq = calculate_codon_freq_normalized(translated_fragments_start_stop)
    # Extract values (frequencies) and keys (elements) from the Counter object
    codon_freq = np.array(list(codon_freq.values()))
    # print(values)

    dicodon_freq = calculate_dicodon_freq_normalized(translated_fragments_start_stop)
    # print(dicodon_freq)

    # sequence_name = get_sequence_name(sequence_file)
    # sequence.id

    return sequence_name, codon_freq, dicodon_freq
```

```python
import os
import glob

directory_path = 'data'
file_list = os.listdir(directory_path)

all_sequences_results = []
all_names_of_sequnces = []

for file in file_list:
    print(file)
    # # Get the current working directory
    # current_directory = os.getcwd()
    # # Print the current working directory
    # print("Current Working Directory:", current_directory)
    file_path = os.path.join(directory_path, file)
    seq_results = calculate_frequencies(file_path)
    all_sequences_results.append(seq_results)
    all_names_of_sequnces.append(seq_results[0])

print(all_sequences_results)
print(all_names_of_sequnces)


# all_sequences_results is a list of tuples ("fasta sequence name", codon_frequencies, dicodon frequencies)
```

```python
def codon_deviation(codon_ferequencies):
    """
    Compare codon frequencies and rate codons based on variation (standard deviation).

    Parameters:
    codon_frquencies (List of lists): List of codon frequency lists.

    Returns:
    dict: Dictionary with codons as keys and their variation (standard deviation) as values.
    """

    if not codon_ferequencies:
        return {}

    codon_variations = {}
    amino_acids = "ACDEFGHIKLMNPQRSTVWY"
    
    for (i, codon) in enumerate(amino_acids):
        codon_frequencies = [codon_freq[i] for codon_freq in codon_ferequencies]
        codon_variations[codon] = np.around(np.std(codon_frequencies), 4)

    return codon_variations
```

```python
codon_deviations = codon_deviation(codon_results)
sorted_items = sorted(codon_deviations.items(), key=lambda item: item[1], reverse=True)
print(sorted_items)
```

```python
def dicodon_deviation(dicodon_matrices):
    """
    Compare dicodon matrices and rate dicodons based on variation (standard deviation).

    Parameters:
    dicodon_matrices (list of 2D arrays): List of dicodon matrices.

    Returns:
    dict: Dictionary with dicodons as keys and their variation (standard deviation) as values.
    """

    if not dicodon_matrices:
        return {}

    dicodon_variations = {}
    amino_acids = "ACDEFGHIKLMNPQRSTVWY"
    
    for i in range(dicodon_matrices[0].shape[0]):
        for j in range(dicodon_matrices[0].shape[1]):
            dicodon = f"{amino_acids[i]}{amino_acids[j]}"

            freqs = [matrix[i, j] for matrix in dicodon_matrices]
            dicodon_variations[dicodon] = np.around(np.std(freqs), 4)

    return dicodon_variations
```

```python
dicodon_deviations = dicodon_deviation(dicodon_results)
sorted_items = sorted(dicodon_deviations.items(), key=lambda item: item[1], reverse=True)
print(sorted_items)
```

```python
def transform_to_phylip(data, names):
    # Number of sequences
    num_sequences = len(data)

    # Print the number of sequences
    print(num_sequences)

    # Print the sequences in Phylip format
    for i in range(num_sequences):
        sequence_name = names[i]
        sequence_data = data[i]
        
        # Join the sequence data into a space-separated string
        sequence_str = " ".join(map(str, sequence_data))
        
        # Print the sequence name and data in Phylip format
        print(f"{sequence_name} {sequence_str}")
```
